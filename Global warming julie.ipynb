```python
import pandas as pd

# Download latest version
#path = kagglehub.dataset_download("ankushpanday1/global-warming-dataset-195-countries-1900-2023")

#print("Path to dataset files:", path)

df = pd.read_csv("./data/global_warming_dataset.csv")
df.head(10)
```

```python
decimal_dict = {"Temperature_Anomaly" : 1 ,"CO2_Emissions" : 1 ,"Population" : 0 ,"Forest_Area" : 1, 
                "Renewable_Energy_Usage" : 2, "Methane_Emissions" : 1, "Sea_Level_Rise" : 2, 'Arctic_Ice_Extent' : 3 , 'Urbanization' :1 ,
       'Deforestation_Rate' : 2, 'Extreme_Weather_Events' : 2, 'Average_Rainfall' : 2,
       'Solar_Energy_Potential' : 2, 'Waste_Management' : 2, 'Per_Capita_Emissions' : 2,'Industrial_Activity' : 2 , 'Air_Pollution_Index' : 2 , 'Biodiversity_Index' : 2,
       'Ocean_Acidification' : 2, 'Fossil_Fuel_Usage' : 2,'Energy_Consumption_Per_Capita' : 2, 'Policy_Score' : 2, 'Average_Temperature' : 2, "GDP" : 9 }

def round_decimal_columns(df, decimal_dict):
    for col, decimals in decimal_dict.items():
        if col in df.columns:
            df[col] = round(df[col],decimals)
    return df
df_rounded = round_decimal_columns(df,decimal_dict)
df_rounded.head(5)
```

```python
numerical_features = df.select_dtypes(include="number").columns.drop("Year")
df1 = df.copy()
df1= df1.groupby(["Year","Country"])[numerical_features].mean().reset_index().sort_values(by="Year", ascending = False).reset_index(drop=True)
df1 
```
